### ALS Model

Let's start with  train/test split.

```python
# Split data into training and test set
(train, test) = sold.randomSplit([.8, .2], seed=41916)
```

Make sure shapes make sense.

```python
print(train.count(), len(train.columns))
```

```python
# Evaluate the model by computing the RMSE on the test data
eval_reg = RegressionEvaluator(metricName="rmse"
                               , labelCol="bought"
                               , predictionCol="prediction")
```

```python
def train_ALS(train, test, num_iters, reg_params, ranks, alphas):
    """
    Grid Search Function to select the best model based on RMSE of hold-out data
    """
    
    # initial
    min_rmse = float('inf')
    best_rank = -1
    best_regularization = 0
    best_alpha = 1
    best_model = None
    
    # tuple up the lists
    combos = [num_iters, reg_params, ranks, alphas]
    combos_tup = list(itertools.product(*combos))
    
    for tup in combos_tup:
    
#     for rank in ranks:
#         for reg in reg_param:
#             for alpha in alphas:
# #                 print("Rank: {}\nReg: {}\nAlpha: {}".format(rank, reg, alpha))
        num_iter = tup[0]
        reg = tup[1]
        rank = tup[2]
        alpha = tup[3]

        # train ALS model
        als = ALS(maxIter=num_iter,
              rank=rank,
              userCol='account_id',
              itemCol='comic_id',
              ratingCol='bought',
              implicitPrefs=True,
              regParam=reg,
              alpha=alpha,
              coldStartStrategy='drop', #Just for CV
              seed=41916)

        model = als.fit(train)

        # Generate predictions on Test
        predictions = model.transform(test)
        predictions.persist()

        rmse = eval_reg.evaluate(predictions)

        print('{} iterations, '.format(num_iter) + 
              '{} latent factors, regularization='.format(rank) +
              '{}, and alpha @ {} : '.format(reg, alpha) +
              'validation RMSE is {:.4f}'.format(rmse))

        if rmse < min_rmse:
            best_rank = rank
            best_regularization = reg
            best_alpha = alpha
            best_model = model

    return best_model
```

### Round 1

```python
# hyper-param config
num_iterations = [15, 20, 30]
ranks = [5, 10, 15, 20]
reg_params = [0.01, 0.1, 0.5]
alphas = [1, 30, 40]
```

```python
# grid search and select best model
start_time = time.time()
final_model = train_ALS(train, test, num_iterations, reg_params, ranks, alphas)

print ('Total Runtime: {:.2f} seconds'.format(time.time() - start_time))
```

Crashed, but on inspection appears that `alpha` at 40 is best. Let's fix it at 40 and pare down some of the grid.

## Results 
Looks like the best parameters we could find are:
- `maxIter` = 25
- `rank` = 10
- `regParam` = 0.1 (default)
- `alpha` = 40

Let's create fit our candidate model.

```python
# Create ALS instance and fit model
als = ALS(maxIter=25,
          rank=10,
          userCol='account_id',
          itemCol='comic_id',
          ratingCol='bought',
          implicitPrefs=True,
          regParam=0.1,
          alpha=40,
          coldStartStrategy='nan', # In the final we want to find our nan's
          seed=41916)
model_use = als.fit(train)
```
